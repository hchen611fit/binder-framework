```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
day4 = pd.read_csv('day4.csv')
x = day4['avgtotalsavings']
y = day4['jobcreated']
plt.scatter(x, y, alpha=0.5)
plt.xlabel('Average Total Savings') 
plt.ylabel('Job Created') 
# displaying the title
plt.title("Day 5 Figure 1")
plt.show()
```

```python
plt.scatter(x, y, alpha=0.5)
plt.xlabel('Average Total Savings') 
plt.ylabel('Job Created') 
plt.xscale('log')
# displaying the title
plt.title('Day 5 Figure 2')
plt.show()
```

```python
plt.hist(np.log10(x), 20, density=1, facecolor='g', alpha=0.75)
plt.xlabel('Logarithm of Average Total Savings')
plt.title('Day 5 Figure 3')
plt.show()
```

```python
df = pd.read_csv('Value_of_Energy_Cost_Savings_Program_Savings_for_Businesses_-_FY2020.csv')
df['month'] = df['Effective Date'].str.split('/',1).str[0]
df_new_1 =df.groupby('month', as_index=False)[['Job created']].sum()
x = df_new_1['month']
y = df_new_1['Job created']
plt.plot(x,y)
plt.xlabel('Month')
plt.ylabel('Job Created') 
plt.title('Day5 Figure 4')
plt.show()
```
